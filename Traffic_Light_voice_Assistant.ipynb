```python
import torch
import cv2
import matplotlib.pyplot as plt
from pathlib import Path
from google.colab.patches import cv2_imshow
import numpy as np

names =  {0: 'person', 1: 'bicycle', 2: 'car', 3: 'motorcycle', 4: 'airplane', 5: 'bus', 6: 'train', 7: 'truck', 8: 'boat', 9: 'traffic light', 10: 'fire hydrant', 11: 'stop sign', 12: 'parking meter', 13: 'bench', 14: 'bird', 15: 'cat', 16: 'dog', 17: 'horse', 18: 'sheep', 19: 'cow', 20: 'elephant', 21: 'bear', 22: 'zebra', 23: 'giraffe', 24: 'backpack', 25: 'umbrella', 26: 'handbag', 27: 'tie', 28: 'suitcase', 29: 'frisbee', 30: 'skis', 31: 'snowboard', 32: 'sports ball', 33: 'kite', 34: 'baseball bat', 35: 'baseball glove', 36: 'skateboard', 37: 'surfboard', 38: 'tennis racket', 39: 'bottle', 40: 'wine glass', 41: 'cup', 42: 'fork', 43: 'knife', 44: 'spoon', 45: 'bowl', 46: 'banana', 47: 'apple', 48: 'sandwich', 49: 'orange', 50: 'broccoli', 51: 'carrot', 52: 'hot dog', 53: 'pizza', 54: 'donut', 55: 'cake', 56: 'chair', 57: 'couch', 58: 'potted plant', 59: 'bed', 60: 'dining table', 61: 'toilet', 62: 'tv', 63: 'laptop', 64: 'mouse', 65: 'remote', 66: 'keyboard', 67: 'cell phone', 68: 'microwave', 69: 'oven', 70: 'toaster', 71: 'sink', 72: 'refrigerator', 73: 'book', 74: 'clock', 75: 'vase', 76: 'scissors', 77: 'teddy bear', 78: 'hair drier', 79: 'toothbrush'}

# Function to get the traffic light color in the current frame
def get_dominant_color(image):
    # Convert image to HSV color space
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Define color ranges in HSV space
    color_ranges = {
        'red': ((0, 100, 100), (10, 255, 255)),  # Low red range
        'green': ((35, 100, 100), (85, 255, 255)),  # Green range
        'yellow': ((20, 100, 100), (30, 255, 255))  # Yellow range
    }

    dominant_color = 'unknown'
    max_area = 0

    for color, (lower, upper) in color_ranges.items():
        mask = cv2.inRange(hsv_image, np.array(lower), np.array(upper))
        area = np.sum(mask > 0)
        if area > max_area:
            max_area = area
            dominant_color = color

    return dominant_color

# Function to process and display the results
def process_frame(frame):
    results = model(frame)  # Perform detection
    a = results[0].boxes.data
    a = a.detach().cpu().numpy()
    px = pd.DataFrame(a).astype("float")
    px.columns = ['xmin', 'ymin', 'xmax', 'ymax', 'confidence', 'class']
    flag = None # a flag indecatior to get a new color only if it changed since the last frame
    for _, row in px.iterrows():
        xmin, ymin, xmax, ymax, confidence, class_id = row[['xmin', 'ymin', 'xmax', 'ymax', 'confidence', 'class']]
        label = names[int(class_id)]  # Get class name
        if label == 'traffic light':  # Get only traffic lights
            roi = frame[int(ymin):int(ymax), int(xmin):int(xmax)]
            color = get_dominant_color(roi)
            print(label, color)
            color_bgr = (0, 255, 0) if color == 'green' else (0, 0, 255) if color == 'red' else (0, 255, 255)
            cv2.rectangle(frame, (int(xmin), int(ymin)), (int(xmax), int(ymax)), color_bgr, 2)
            cv2.putText(frame, color.capitalize(), (int(xmin), int(ymin) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color_bgr, 2)
            if color == 'green' and flag != 'green':
              voice_timestamps.append((2,count / fps))
            elif color == 'yellow' and flag != 'yellow':
              voice_timestamps.append((1,count / fps))
            else:
              if flag != 'red':
                voice_timestamps.append((0,count / fps))
            flag = color
            break # if there are more then one traffic light, break the loop and move on to the next frame
    return frame

# Load video
video_path = 'traffic_light.mp4'
cap = cv2.VideoCapture(video_path)
count = 0
fps = cap.get(cv2.CAP_PROP_FPS)  # Get frames per second
voice_timestamps = []


while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    count += 1
    processed_frame = process_frame(frame)

    # Display the resulting frame
    if count % 100 == 0:
      cv2_imshow(processed_frame)

    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# Load the video without sound
video = VideoFileClip("traffic_light.mp4")

# Load the sounds
beep_clip_red = AudioFileClip("red.wav")
beep_clip_green = AudioFileClip("green.wav")
beep_clip_yellow = AudioFileClip("yellow.wav")


# Create a list of sounds at specific timestamps
beep_clips = []
test = []
begin = None
for t in voice_timestamps:
    curr = t[0]

    if curr == 0 and (begin == None or begin == 'Y'):
        beep_clip = beep_clip_red
        beep_clips.append(beep_clip.set_start(t[1]).set_duration(beep_clip.duration))
        begin = 'R'
        test.append(begin)
    elif curr == 1 and (begin == None or begin == 'G'):
        beep_clip = beep_clip_yellow
        beep_clips.append(beep_clip.set_start(t[1]).set_duration(beep_clip.duration))
        begin = 'Y'
        test.append(begin)

    elif curr == 2 and (begin == None or begin == 'R'):
        beep_clip = beep_clip_green
        beep_clips.append(beep_clip.set_start(t[1]).set_duration(beep_clip.duration))
        begin = 'G'
        test.append(begin)






# Combine all sounds into a single audio track
final_audio = CompositeAudioClip(beep_clips)

# Add the final audio to the video
final_video = video.set_audio(final_audio)

# Save the final video with sound
final_video.write_videofile("trafic_light_with_sound.avi", codec="libx264", audio_codec="aac")
```
